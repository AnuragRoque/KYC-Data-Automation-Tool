```python
import re
import csv
```

```python
def extract_info(filepath):
    try:
        doc = fitz.open(filepath)
        text = doc[0].get_text()  # Extract text from the first page
        doc.close()
        # print(text)
        # Extract information using regular expressions
        if 'Permanent' in text:
            pan = re.findall(r"[a-zA-Z]{5}[0-9]{4}[A-Za-z]", text)
            doi=re.findall(r"[0-9]{2}/[0-9]{2}/[0-9]{4}",text)
            try:
                name=re.findall(r"(.+?)\n(.+?)\n[0-9]{2}/[0-9]{2}/[0-9]{4}",text)
            except AttributeError:
                name=None
            
            
            info={'doi':doi,'name':name,'pan':pan,'path':filepath}
            print(info)
            return info
            
    except Exception as e:
        print(f"Error processing file {filepath}: {e}")
    return None

def main():
    csv_folder_name = r"C:\Users\trpwkycaudit1\Documents\Validation-data\path1.csv"
    pan_cid_info = []

    with open(csv_folder_name, 'r', newline='',encoding='UTF-8') as csvfile:
        reader = csv.reader(csvfile)
        next(reader)  # Skip header
        for row in reader:
            filepath = row[0]
            info = extract_info(filepath)
            if info:
                pan_cid_info.append(info)
                
    print(pan_cid_info)
    
    if pan_cid_info:
    
        csv_file_path = "zx-unique-pan-files-cid.csv"
        with open(csv_file_path, "w", newline='',encoding='UTF-8') as csvfile:
            writer = csv.DictWriter(csvfile, fieldnames=['doi','name','pan','path'])
            writer.writeheader()
            for item in pan_cid_info:
                writer.writerow(item)
        
        print("Data written to", csv_file_path)
    
    

if __name__ == "__main__":
    main()
    
```
