```python
import re

# Function to extract information from text
def extract_info_from_text(text):
    # Initialize variables to store extracted information
    name = None
    fname = None
    dob = None
    gender = None
    address = None
    vid_aadhaar = None

    # Split text into lines
    lines = text.split("\n")

    # Iterate through each line
    for line in lines:
        # Check for name
        if "Fname" in line:
            # Assuming the name follows "Fname"
            name = line.split("Fname")[-1].strip()
        
        # Check for father's name
        if "S/O" in line or "D/O" in line:
            fname = line.split("S/O")[-1].split("D/O")[-1].strip()
        
        # Check for DOB
        if "DOB" in line or "YOB" in line:
            dob = re.search(r'\d{1,2}/\d{1,2}/\d{4}', line).group(0)

        # Check for gender
        if "MALE" in line or "FEMALE" in line:
            gender = line.strip()

        # Check for address
        if "ADDRESS" in line:
            # Assuming address starts after "ADDRESS"
            address_lines = []
            address_index = lines.index(line) + 1
            while address_index < len(lines):
                if re.match(r'^\d{6}$', lines[address_index].strip()):
                    # Assuming pincode is a 6-digit number
                    break
                address_lines.append(lines[address_index])
                address_index += 1
            address = "\n".join(address_lines).strip()

        # Check for VID/AADHAAR
        if "VID" in line or "AADHAAR" in line:
            # Assuming VID/AADHAAR follows "VID" or "AADHAAR"
            vid_aadhaar = re.search(r'\d{4}\s+\d{4}\s+\d{4}\s+\d{4}', line).group(0)

    # Return extracted information
    return {
        "Name": name,
        "Father's Name": fname,
        "DOB": dob,
        "Gender": gender,
        "Address": address,
        "VID/AADHAAR": vid_aadhaar
    }

# Sample text data
sample_text = """
Random trash data
SHAILAZ NAG
S/O LATE.S.N. NAG
FLAT NO-D3-902
CLARION THE LEGEND
SECTOR-57
SAMASPUR
GURGAON SECTOR 45
GURGAON HARYANA- 122003
9871579827
.....
INFORMATION
THIS IS ELECTRONICALLY GENERATED LETTER.
- 
7865
VID 9194 4806 0620 3901
"""

# Extract information from the sample text
extracted_info = extract_info_from_text(sample_text)

# Print extracted information
for key, value in extracted_info.items():
    print(f"{key}: {value}")
```
